```python
import pandas as pd
import warnings
import nameparser
warnings.filterwarnings('ignore')



```

```python
trans = pd.read_csv('../data/transactions_20230914.csv')

trans['disclosure_date'] = pd.to_datetime(trans['disclosure_date'])
trans['transaction_date'] = pd.to_datetime(trans['transaction_date'])

trans['parsed_name'] = trans['member'].apply(lambda x: nameparser.HumanName(x))
trans['first_name'] = trans['parsed_name'].apply(lambda x: x.first)
trans['last_name'] = trans['parsed_name'].apply(lambda x: x.last)
trans_117 = trans[(trans['transaction_date'] < '2023-01-03') & (trans['transaction_date']  >= '2021-01-03')]
trans_116 = trans[(trans['transaction_date'] < '2021-01-03')]
trans_118 = trans[(trans['transaction_date'] > '2023-01-03')]
trans_116.shape, trans_117.shape, trans_118.shape
```

```python
h_assignments = pd.read_excel('../data/unprocessed files/House of Representative Standing Committee Assignments 103rd-117th/House_assignments_103-117.xls',
                              skiprows=1)


h_assignments['parsed_name'] = h_assignments['Name'].apply(lambda x: nameparser.HumanName(x))
h_assignments['first_name'] = h_assignments['parsed_name'].apply(lambda x: x.first)
h_assignments['last_name'] = h_assignments['parsed_name'].apply(lambda x: x.last)
h_assignments_117 = h_assignments[h_assignments['Congress'] == 117]
h_assignments_116 = h_assignments[h_assignments['Congress'] == 116]
h_assignments_117
```

```python
merged_117 = trans_117.merge(h_assignments_117, on =['first_name', 'last_name'], how='left')
merged_116 = trans_116.merge(h_assignments_116, on =['first_name', 'last_name'], how='left')
merged_117
```
